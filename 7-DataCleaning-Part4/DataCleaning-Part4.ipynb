```python
week1.head(2)
```

```python
week2.head(2)
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week1 ", " - Week2"])
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - Week1 ", " - Week2"]).count()
```

```python
week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["- Week1", " - Week2"],
            indicator = True)
```

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = ["- Week1", " - Week2"],
            indicator = True)
```
